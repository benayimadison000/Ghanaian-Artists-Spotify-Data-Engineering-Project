```python
final_analysis = spark.read.parquet("/mnt/musicstg/gold/artists_final/")
```

```python
display(final_analysis)
```

```python
from pyspark.sql.functions import explode, col

# Explode the 'genres' array column so each genre gets its own row, keeping artist_id and artist_name
artist_genres = (
    final_analysis
    .select(
        "artist_id",
        "artist_name",
        explode("genres").alias("genre")
    )
)
# Display the resulting DataFrame with one row per artist-genre pair
display(artist_genres)
```
