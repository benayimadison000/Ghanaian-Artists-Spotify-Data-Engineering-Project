```python
genres = ["pop", "hip hop", "afrobeats", "rock", "latin", "country"]  # List of music genres to search for artists

all_artists = []  # Initialize an empty list to store all artist data

for g in genres:  # Iterate over each genre in the genres list
    params = {  # Define the parameters for the Spotify API search request
        "q": f"genre:{g}",  # Set the search query to filter by the current genre
        "type": "artist",  # Specify that we are searching for artists
        "limit": 50  # Limit the number of results to 50 artists per genre
    }

    response = requests.get(  # Send a GET request to the Spotify API search endpoint
        "https://api.spotify.com/v1/search",  # Spotify API endpoint for searching
        headers={"Authorization": f"Bearer {token}"},  # Include the access token in the request headers for authentication
        params=params  # Pass the search parameters defined above
    )

    data = response.json()  # Parse the API response as JSON and store it in the 'data' variable
    all_artists.extend(data["artists"]["items"])  # Add the list of artist items from the response to the all_artists list
```

```python
top_artists = sorted(all_artists, key=lambda x: x["popularity"], reverse=True)
top_100 = top_artists[:100]
```
