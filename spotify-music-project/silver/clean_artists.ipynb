```python
bronze_df = spark.read.json("/mnt/musicstg/bronze/artists_full/")
```

```python
bronze_df.printSchema()
```

```python
artists_silver = (
    bronze_df
    # Select relevant artist fields from the nested JSON structure
    .select(
        "artist_id",
        "artist.name",
        "artist.genres",
        "artist.popularity",
        "artist.followers.total",
        "artist.external_urls.spotify",
        "artist.images"
    )
    # Rename columns for clarity
    .withColumnRenamed("name", "artist_name")
    .withColumnRenamed("total", "followers")
    # Remove duplicate artists based on artist_id
    .dropDuplicates(["artist_id"])
)
```

```python
 display(artists_silver)
```

```python
artists_silver.write.mode("overwrite").parquet("/mnt/musicstg/silver/artists/")
```
